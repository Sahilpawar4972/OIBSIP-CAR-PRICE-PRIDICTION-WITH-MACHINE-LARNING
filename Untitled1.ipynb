```python
dataset = dataset[dataset['year'].str.isnumeric()]
```

```python
dataset['year'] = dataset['year'].astype(int)
```

```python
dataset.info()
```

```python
dataset.shape
```

```python

dataset[dataset['Price'] == 'Ask For Price'].count()
```

```python
dataset = dataset[dataset['Price'] != 'Ask For Price']
```

```python
dataset.shape
```

```python
dataset['Price'] = dataset['Price'].str.replace(',', '').astype(int)
```

```python
dataset.info()
```

```python
dataset['kms_driven'] = dataset['kms_driven'].str.split(' ').str.get(0).str.replace(',', '')
dataset['kms_driven'].unique()
```

```python
dataset = dataset[dataset['kms_driven'].str.isnumeric()]
```

```python
dataset.shape
```

```python
dataset['kms_driven'] = dataset['kms_driven'].astype(int)
```

```python
dataset.info()
```

```python
dataset  = dataset[~dataset['fuel_type'].isna()]
```

```python
dataset.info()
```

```python
dataset = dataset[dataset['Price']<6e6].reset_index(drop = True)
```

```python
dataset.describe()
```

```python
dataset.to_csv('cleaned_car_dataset.csv')
```

```python
X = dataset.drop(columns = 'Price')
y = dataset['Price']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder

from sklearn.compose  import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(X[['name', 'company', 'fuel_type']])
```

```python
ohe.categories_
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories = ohe.categories_), ['name', 'company', 'fuel_type']), remainder = 'passthrough')
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans, lr)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```

```python
scores=[]

for i in range(1000):
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state= i)
  lr = LinearRegression()
  pipe = make_pipeline(column_trans, lr)
  pipe.fit(X_train, y_train)
  y_pred = pipe.predict(X_test)
  print(r2_score(y_test, y_pred), i)
  scores.append(r2_score(y_test, y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```
